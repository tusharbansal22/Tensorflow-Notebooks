```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,20,1000)
```

```python
y=np.sin(x)+x*0.2
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('input')
plt.ylabel('output')
plt.title('my plot');
```

## Scatterplot

```python
X=np.random.randn(100,2)
plt.scatter(X[:,0],X[:,1])
```

```python
X=np.random.randn(200,2)
X[:50]+=3
Y=np.zeros(200)
Y[:50]=1
plt.scatter(X[:,0],X[:,1],c=Y);
```
